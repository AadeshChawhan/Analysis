## Countries listed According to their Quantity sale

```python
df
```

```python
df_without_uk = df[df['Country'] != 'United Kingdom']

country_demand = df_without_uk.groupby('Country')['Quantity'].sum().reset_index()
country_demand_sorted = country_demand.sort_values(by = 'Quantity', ascending = False)

country_demand_sorted
```

```python
plt.figure(figsize=(12,8))
# Since the heatmap needs a matrix, we pivot the data to create a DataFrame
country_demand_pivot = country_demand.set_index('Country').T

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(country_demand_pivot, annot=True, fmt='.0f', cmap='YlGnBu', cbar=True, linewidths=.5)

plt.title('Product Demand by Country (Excluding United Kingdom)')
plt.xlabel('Country')
plt.ylabel('Demand')

plt.show()
```
